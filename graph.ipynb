Here i use 

$$\tag{1}
    f_{\pm}=\dfrac{1}{2}(f_{SRR}+f_{YIG})\pm\dfrac{1}{2}\sqrt{(f_{SRR}-f_{YIG})^2+\delta^2}
$$

With $f_{YIG}$ is in FMR mode which is well fitted by the Kittel formula,

$$\tag{2}
    f=\gamma\sqrt{(H+H_0)(H+H_0+4\pi M_s)}
$$

I'm using YIG as my sample with a saturation magnetization value of $4\pi M_s=1750$ G, anisotropic field $H_0=0$ Oe and gyromagnetic ratio $\gamma=2.72$ MHz Oe$^{-1}$.$^{[1]}$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
#Declare the value of your variables
sat = 1750
aniso = 0
gyro = 2.72e+06
fsrr = 2.4e+09
cop = 65.7e+06
#this is the H
x = np.linspace(0,1000,1000)
#f_YIG using Equation (2)
f = gyro*np.sqrt((x+aniso)*(x+sat+aniso))
#y1 is the upper mode and y2 is the lower mode
y1 = (1/2)*(fsrr+f)+(1/2)*np.sqrt((fsrr-f)**2+(cop)**2)
y2 = (1/2)*(fsrr+f)-(1/2)*np.sqrt((fsrr-f)**2+(cop)**2)

#Setting up the plot
fig, ax = plt.subplots()
ax.plot(x,y1, x,y2)
ax.set_title('Magnon-Polariton Anti-crossing Curve')
plt.xlabel('Magnetic field (Oe)')
plt.ylabel('Frequency (Hz)')
plt.xlim(200, 550)
plt.ylim(2e+09, 2.8e+09)
```
